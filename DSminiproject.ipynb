```python
!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
keywords = ["Data Science", "Machine Learning", "Artificial Intelligence"]
trends.build_payload(keywords, timeframe='today 5-y')
data = trends.interest_over_time().drop(columns=['isPartial'])
data.head()
#Histogram
data.plot(kind='hist', bins=20, alpha=0.7, figsize=(10, 6))
plt.title("Histogram of Google Search Interest")
plt.xlabel("Interest Level")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=data) # Now sns is defined and can be used
plt.title("Box Plot of Google Search Interest Over Time")
plt.ylabel("Interest Level")
plt.show()
```

```python
#Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data.index, data['Data Science'], label="Data Science", alpha=0.5)
plt.scatter(data.index, data['Machine Learning'], label="Machine Learning", alpha=0.5)
plt.scatter(data.index, data['Artificial Intelligence'], label="Artificial Intelligence", alpha=0.5)
plt.title("Scatter Plot of Search Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Interest Level")
plt.legend()
plt.show()
```

```python
#Pearson Correlation
correlation = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()
#Explanation:
# A value closer to 1 indicates a strong positive relationship, while values closer to -1 indicate a strong negative relationship.
# For example, a high positive correlation between "data science" and "machine learning" would indicate that searches for these terms trend together.

```

 Identify Dependent and Independent Features
1.In this case, interest in each topic is independent, as they dont rely on each other directly.
2.To analyze relationships, consider each topic as a separate variable of interest.

```python
# Plot trend for each keyword to analyze changes over time
plt.figure(figsize=(14, 8))
for column in data.columns:
    plt.plot(data.index, data[column], label=column)
plt.title("Google Search Interest Over Time")
plt.xlabel("Date")
plt.ylabel("Interest Level")
plt.legend(title="Keywords")
plt.grid()
plt.show()
```
